The Pandas documentation [here](https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html#applying-multiple-functions-at-once) should be a useful reference.

```
top_nationalities = df['nationality'].value_counts().head(15).index
top_nationalities_df = df[df['nationality'].isin(top_nationalities)]
```

```
stacked_bar_chart = alt.Chart(top_nationalities_df).mark_bar().encode(
    x=alt.X('nationality:N', title='Nationality'),
    y=alt.Y('count():Q', title='Number of Players'),
    color=alt.Color('preferred_foot:N', title='Preferred Foot', legend=alt.Legend(orient='right')),
    tooltip=['nationality', 'preferred_foot', 'count()']
).properties(
    title="Top 15 Nationalities By Number of Players Stacked By Preferred Foot",
    width=600,
    height=600
)

stacked_bar_chart
```
