```
df = pd.read_json("tweet.json")

df['created_at'] = df['tweet'].apply(lambda x: x.get("created_at"))

df["created_at"] = pd.to_datetime(df["created_at"])

df["day_of_week"] = df['created_at'].dt.day_name()
df["hour"] = df["created_at"].dt.hour

group_by_time_df = df.groupby(["day_of_week", "hour"]).size().reset_index(name="count")

group_by_time_df.head()
```

```
heatmap = alt.Chart(group_by_time_df).mark_rect().encode(
    x=alt.X('day_of_week:O', title="Day of the Week"),
    y=alt.Y('hour:O', title="Time of Day"),
    color='count',
    tooltip=["day_of_week", "hour", "count"]
).properties(
    width=500,
    title="Frequency of Tweets by Day and Hour"
)

heatmap.show()
```
