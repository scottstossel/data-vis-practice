```
nationality_df = df.groupby('nationality').agg(
    number_of_players=('sofifa_id', 'size'),
    average_age = ('age', 'mean')
)
nationality_df
```

```
import altair as alt
!pip install "vl-convert-python>=1.6.0"
!pip install "vegafusion[embed]>=1.5.0"
alt.data_transformers.enable("vegafusion")
```

```
wage = alt.Chart(df).mark_point().encode(
    x=alt.X('wage_eur:Q', title="Wage (Euros)", axis=alt.Axis(
       titleFontSize=16,
       labelFontSize=14
    ), scale=alt.Scale(zero=False)),
    color=alt.Color('age:Q', scale=alt.Scale(scheme='redblue')),
    y=alt.Y('potential:Q', axis=alt.Axis(
      titleFontSize=16,
      labelFontSize=14
    ), title='Player Potential', scale=alt.Scale(domain=[df.potential.min(),100])),
    tooltip=['short_name', 'wage_eur', 'potential', 'pace']
  ).properties(
    title='Soccer Player Wage vs Potential',
    height=500,
    width=400
  ).interactive()
```

```
pace = alt.Chart(df).mark_point().encode(
    x=alt.X('pace:Q', title="Pace", axis=alt.Axis(
       titleFontSize=16,
       labelFontSize=14
    ), scale=alt.Scale(zero=False)),
    color=alt.Color('age:Q', scale=alt.Scale(scheme='redblue')),
    y=alt.Y('potential:Q', axis=alt.Axis(
      titleFontSize=16,
      labelFontSize=14
    ), title='Player Potential', scale=alt.Scale(domain=[df.potential.min(),100])),
    tooltip=['short_name', 'wage_eur', 'potential', 'pace']
  ).properties(
    title='Soccer Player Pace vs Potential',
    height=500,
    width=400
  ).interactive()
```
