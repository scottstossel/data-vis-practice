```
import altair as alt
```

```
from vega_datasets import data

driving = data.driving()
driving.head()
```

```
year_min = driving['year'].min()
year_max = driving['year'].max()

alt.Chart(driving).mark_point().encode(
    x=alt.X('year:Q', scale=alt.Scale(domain=[year_min, year_max])),
    y='miles',
).properties(title='Year vs Miles Driven')
```
